### Задание 4.3.2

```python
def f(x):
    y = (sqrt(1 + e**sqrt(x) + cos(x**2)) / abs(1 - sin(x)**3)) + log(abs(2*x))
    return y

x = [i for i in range(1, 10 + 1)]
y = [f(i) for i in range(1, 10 + 1)]

slice_x = x[:len(x)//2]
slice_y = y[:len(y)//2]

fig = plt.figure(figsize=(10, 7))
plt.plot(x, y)
plt.scatter(slice_x, slice_y)
```

### Задание 4.3.3

```python
def f(x):
    y = abs(cos(x * e**(cos(x) + log(x+1))))
    return y

x = np.arange(0.0, 10, 1.0)
y = [f(i) for i in x]

plt.grid()
plt.plot(x, y, c = 'r')
plt.fill_between(x, y)

area = trapz(y)
print(area)

```
